Dicho esto, un archivo CSV es cualquier archivo de texto en el cual los caracteres están separados por comas, haciendo una especie de tabla en filas y columnas. En pandas, el método **read_csv()** puede leer archivos con columnas separadas por comas en un DataFrame de pandas. La estructura que sigue es la siguiente:

Primero importamos pandas

````python
import pandas as pd 
````

Después definimos una variable con la ruta del archivo

````python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
````

Y después usamos el método read_csv para importar los datos

````python
df=pd.read_csv(url)
````

El método **pd.read_csv** asume que el archivo tiene un encabezado como se muestra a continuación:

````python
df.head()
````

Para corregir esto dentro de los parámetros de **pd.read_csv** usamos **header == None**

````python
df=pd.read_csv(url,header=None)
````

## Imprimir el dataframe en Python
* **df** imprime el dataframe completo (no recomendado para datasets grandes**)
* **df.head(n)** muestra las primeras *n* filas del dataframe.
* **df.tail(n)** muestra las últimas *n* filas del dataframe.

Usamos df.head() para visaulizar el dataframe

````python
df.head()
````
