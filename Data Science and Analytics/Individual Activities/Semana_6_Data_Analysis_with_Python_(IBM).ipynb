````python
import pandas as pd
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df=pd.read_csv(url, names = headers)
df=df.fillna(df.mode().iloc[0])
````

# GroupBy en Python
El método **groupby** es usado en variables categóricas, lo que hace es que agrupa los datos en subconjuntos de acuerdo con las diferentes categorías de las variables. 

En resumen este método hace lo siguiente:
* Puede aplicarse en variables categóricas
* Agrupar datos en categorías
* Variables únicas o múltiples

Por ejemplo si queremos saber el precio promedio de un vehículo y observar como difieren entre diferentes las variables como línea de carrocería *body-style* y tracción *drive-wheels**.

Para hacer esto seleccionamos las columnas que nos interesan y las guardamos dentro de otro **dataframe**

````python
df_test=df[["drive-wheels","body-style","price"]]
````

Después agrupamos los datos reducidos de acuerdo con *drive-wheels* y *body-style*, dado que nos interesa conocer el precio promedio, usamos el método **mean()** de pandas.

````python
df_grp=df_test.groupby(["drive-wheels","body-style"],as_index=False).mean()
df_grp
````

Esta forma de visualizar datos no es la más optima, por ello podemos usar una tabla pivot para tener una variable desplegada dentro de las columnas y otras variable desplegada en las filas.

## Método de pandas - Pyvot()
- Una variable se muestra en las columnas y la otra variable se muestra en las filas. Para visualizar la tabla en formato pivot simplemente hay que usar el método *Pyvot()* dentro de la librería pandas

````python
df_pivot=df_grp.pivot(index="drive-wheels",columns="body-style")
````

## ANOVA

Es un test estadístico que significa **Análisis de la Varianza** 

¿Por qué realizamos ANOVA?
* Se usa para encontrar la correlación entre diferentes grupos de variables categóricas.

¿Qué obtenemos del ANOVA?
* F-test score: variación entre la media del grupo de muestra divido entre la variación dentro del grupo de muestra.
* P-value: muestra si el resultado es estadísticamente significativo

Un score F pequeño indica una correlación pobre entre variables categóricas y la variable objetivo, tal como se puede observar en el siguiente ejemplo:

trdvtrhhyg.PNG

El F-test calcula la proporción de variación entre la media del grupo sobre la variación de cada uno de los grupos de muestra, el diagrama ilustra el caso donde el score del F-test es pequeño. Porque, tal como vemos la variación entre los precios de cada grupo es más larga que la diferencia de los valores promedio de cada grupo.

Ahora veamos el siguiente diagrama:

Como podemos observar el score del F-Test es grande. La variación entre los 2 grupos es comparable a las variaciones entre los dos grupos. Dado que el F-Score es grande, entonces la correlación es fuerte en este caso. De regreso al ejemplo, el gráfico de barras muestra el precio promedio para diferentes categorías de la columna *make*.

ftgr.PNG

Como se puede observar el F-score esperado, entre un carro de la marca Honda y otro de la marca Subaru es pequeño, dado que existe una pequeña diferencia de precios entre las 2 marcas, sin embargo, podemos esperar un F-Value alto si comparamos entre Honda y Jaguar, esto debido a que la diferencia de precios es grande. 

Sin embargo, en el diagrama no podemos conocer exactamente las varianzas, para ello usaremos **ANOVA**.

En la primera línea extraemos las columnas **make** y **price**.

````python
df_anova=df[["make","price"]]
````

Posteriormente agrupamos los datos en base de las diferentes marcas.

````python
grouped_anova=df_anova.groupby(["make"])
````

La prueba Anova puede ser desarrollada en Python usando el método **f_oneway** con la función contruida en el paquete Scipy. Pasamos los datos de *price* a los 2 grupos que queremos comparar y calcula el resultado ANOVA.

````python
from scipy import stats
anova_results_l=stats.f_oneway(grouped_anova.get_group("honda")["price"],grouped_anova.get_group("subaru")["price"])
anova_results_l
````

El resultado confirma lo que se suponía al principio, los precios entre autos de la marca Honda y Subaru no son significativamente diferentes, dado que el score de F-test es menor a 1 y el p-value es mayor a 0.05, podemos hacer lo mismo con Honda y Jaguar.

````python
anova_results_l=stats.f_oneway(grouped_anova.get_group("honda")["price"],grouped_anova.get_group("jaguar")["price"])
anova_results_l
````

El F-Score es muy grande (de 401) y el p-value es mayor a 0.05. Por lo que podemos concluir que existe un correlación fuerte entre una variable categórica y otras variables, si la prueba ANOVA nos da un valor grande de F-test y un p-value bajo.

# Correlación
Es una métrica estadística para medir a que extensión diferentes variables son interdependientes. 

Por ejemplo: 
* El cáncer de pulmón está correlacionado a fumar
* Mayor precipitación en el clima significa un aumento en el uso de paraguas

La correlación no implica causación. Por ejemplo no podemos decir que usar el paraguas haga que llueva o que la lluvia cause que se use el paraguas. En ciencia de datos usualmente se usa la **correlación**.

Observemos la correlación entre dos columnas (engine-size y price), para ello visualizaremos las 2 variables usando scatterplot y una línea llamada **regresión lineal** que indica la relación entre las dos. El objetivo del gráfico es observar si el tamaño del motor (engine-size) tiene algún impacto en el precio. En este ejemplo se puede observar que la línea recta a lo largo de los data points es empinada, lo que indica que hay una relación linear positiva entre 2 variables.


````python
def mi_regplot(x,y): #hice una función para no repetir el código
  sns.regplot(x=x,y=y,data=df,color="royalblue")
  plt.ylim(0,)
````

````python
mi_regplot("engine-size","price")
````

Con el incremento en el tamaño de los valores del tamaño del motor, el valor del precio también sube y la pendiente de la línea es positiva, así que existe una correlación positiva entre **engine-size** y **price**.

# Correlación - Estadísticas
## Correlación de Pearson
* Mide la fuerza de la correlación entre 2 columnas, este método nos dará dos valores:
  * Correlación de coeficientes
  * P-value

* Para el **coeficiente de correlación** tenemos los siguientes casos:
  * Cercano a +1: Relación grande positiva
  * Cercano a -1: Relación grande negativa
  * Cercano a 0: Sin relación
* El **p-value** nos dirá que tan certera es la correlación que hemos calculado.
  * P-value < 0.001 **Fuerte** certeza en el resultado
  * P-value < 0.05 **Moderada** certeza en el resultado
  * P-value < 0.1 **Débil** certeza en el resultado
  * P-value > 0.1 **No hay** certeza en el resultado


Podemos decir que tenemos una correlación fuerte si:
* La correlación es cercana a 1 0 -1
* El P-value es menor que 0.001

El siguiente plot muestra datos con diferentes valores de correlación:

vgybnytny.PNG

Para obtener el coeficiente de pearson y el p-value hacemos lo siguiente:

````python
pearson_coef,p_value=stats.pearsonr(df["horsepower"],df["price"])
````

````python
print(pearson_coef)
print(p_value)
````

**¿Qué es el análisis de la varianza (ANOVA)?**

El análisis de la varianza (ANOVA) es una herramienta de análisis utilizada en estadística que divide la variabilidad agregada observada dentro de un conjunto de datos en dos partes: factores sistemáticos y factores aleatorios. Los factores sistemáticos tienen una influencia estadística en el conjunto de datos dado, mientras que los factores aleatorios no la tienen. Los analistas utilizan la prueba ANOVA para determinar la influencia que tienen las variables independientes en la variable dependiente en un estudio de regresión.
